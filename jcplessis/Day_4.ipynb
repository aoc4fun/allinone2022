```python
def parse_section(sections):
  return [[int(x) for x in section.split("-")] for section in sections.split(",")]

def is_fully_included(contained, container):
  return contained[0] >= container[0] \
    and contained[1] <= container[1]

def count_sections(assignments, filter):
  result = 0
  for sections in assignments.split("\n"):
    section1, section2 = parse_section(sections)
    if filter(section2, section1):
      result += 1
  return result


def count_included_sections(assignments):
  return count_sections(assignments, 
    lambda section1, section2 : is_fully_included(section1, section2) \
      or is_fully_included(section2, section1)
  )

def has_overlap(overlapping, overlapped):
  return overlapping[0] >= overlapped[0] and overlapping[0] <= overlapped[1] \
    or overlapping[1] >= overlapped[0] and overlapping[1] <= overlapped[1]

def count_overlapping_sections(assignments):
  return count_sections(assignments, 
    lambda section1, section2 : has_overlap(section1, section2) \
      or has_overlap(section2, section1)
  )
```

```python
import unittest

example = """2-4,6-8
2-3,4-5
5-7,7-9
2-8,3-7
6-6,4-6
2-6,4-8"""

class TestAcceptance(unittest.TestCase):

    def test_parse_section(self):
      self.assertEqual([[2,4], [6,8]], parse_section("2-4,6-8"))

    
    def test_is_fully_included(self):
      self.assertTrue(is_fully_included([3,7], [2,8]))
      self.assertFalse(is_fully_included([3,12], [2,8]))

    def test_count_included_sections(self):
      self.assertEqual(2, count_included_sections(example))

    def test_has_overlap(self):
      self.assertTrue(has_overlap([3,7], [2,8]))
      self.assertFalse(has_overlap([3,5], [7,8]))

    def test_count_overlapping_sections(self):
      self.assertEqual(4, count_overlapping_sections(example))


unittest.main(argv=[''], verbosity=2, exit=False)
```
